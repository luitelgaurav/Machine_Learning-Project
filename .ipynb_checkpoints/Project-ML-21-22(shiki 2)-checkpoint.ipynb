```python
#Use Chi-square test to filter non metric features
non_metric_features_selected = []

for non_metric in non_metric_features:
    dfObserved = pd.crosstab(df_train["Buy"], df_train[non_metric]) 
    chi2, p, dof, expected = stats.chi2_contingency(dfObserved.values)
    if p < 0.05:
        non_metric_features_selected.append(non_metric)
```

## 5.1. Define functions for running models

```python
#Function for running with Repeated k-fold cross-validation
from sklearn.metrics import f1_score
from sklearn.model_selection import RepeatedKFold
from sklearn.preprocessing import MinMaxScaler

def avg_score_skf_scaled(n_splits,n_repeats,X,y):
    skf = StratifiedKFold(n_splits)
    score_train = []
    score_val = []
    score_f1 = []
    f1_best = 0.0
    for i in range(n_repeats):
        for train_index, val_index in skf.split(X,y):
            
            X_train, X_val = X.iloc[train_index], X.iloc[val_index]
            y_train, y_val = y.iloc[train_index], y.iloc[val_index]
            
            scaler = MinMaxScaler().fit(X_train)
            X_train_scaled = scaler.transform(X_train) 
            X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns).set_index(X_train.index)
            X_val_scaled = scaler.transform(X_val) 
            X_val_scaled = pd.DataFrame(X_val_scaled, columns = X_val.columns).set_index(X_val.index)
        
            model_fit = model.fit(X_train_scaled, y_train)
        
            y_pred = model.predict(X_val_scaled)
        
            value_train = model_fit.score(X_train_scaled, y_train)
            value_val = model_fit.score(X_val_scaled, y_val)
            value_f1 = f1_score(y_val, y_pred)
        
            score_train.append(value_train)
            score_val.append(value_val)
            score_f1.append(value_f1)
            
    return(np.mean(score_train), np.mean(score_val), np.mean(score_f1))
```

```python
#Function for running with Stratified k-fold cross-validation
from sklearn.metrics import f1_score
from sklearn.model_selection import StratifiedKFold

def avg_score_skf(n_splits,n_repeats,X,y):
    skf = StratifiedKFold(n_splits)
    score_train = []
    score_val = []
    score_f1 = []
    f1_best = 0.0
    for i in range(n_repeats):
        for train_index, val_index in skf.split(X,y):
            
            X_train, X_val = X.iloc[train_index], X.iloc[val_index]
            y_train, y_val = y.iloc[train_index], y.iloc[val_index]
            
            model_fit = model.fit(X_train, y_train)
            
            y_pred = model.predict(X_val)
        
            value_train = model_fit.score(X_train, y_train)
            value_val = model_fit.score(X_val,y_val)
            value_f1 = f1_score(y_val, y_pred)
        
            score_train.append(value_train)
            score_val.append(value_val)
            score_f1.append(value_f1)

    return(np.mean(score_train), np.mean(score_val), np.mean(score_f1))
```
