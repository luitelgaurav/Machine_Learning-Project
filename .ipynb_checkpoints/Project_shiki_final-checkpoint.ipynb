## [6.1.Define functions for running models](#M)
    
</a>

```python
def avg_score_skf(n_splits, n_repeats, X, y, model, smote=False):
    
    score_train = []
    score_val = []
    score_f1 = []
    list_scaler = []

    f1_best = 0.0
    for i in range(n_repeats):
        
        skf = StratifiedKFold(n_splits, shuffle=True, random_state=np.random.randint(1000))
        
        for train_index, val_index in skf.split(X,y):
            
            #Split the training and valiation dataset
            X_train, X_val = X.iloc[train_index], X.iloc[val_index]
            y_train, y_val = y.iloc[train_index], y.iloc[val_index]
            
            #Target Encoder
            list_t = list(set(targetlist)&set(feature_selected))
            X_train[list_t] = X_train[list_t].astype(str, copy=False)
            t_encode = TargetEncoder().fit(X_train[list_t], y_train)
            
            X_train[list_t] = t_encode.transform(X_train[list_t])
            X_val[list_t] = t_encode.transform(X_val[list_t])
            
            #Power Transformation
            power = PowerTransformer().fit(X_train)
            X_train = pd.DataFrame(power.transform(X_train), 
                                          columns = X_train.columns).set_index(X_train.index)
            X_val = pd.DataFrame(power.transform(X_val), 
                                        columns = X_val.columns).set_index(X_val.index)

            #MinMaxScaler
            scaler = MinMaxScaler().fit(X_train)
            
            X_train_scaled = pd.DataFrame(scaler.transform(X_train), 
                                          columns = X_train.columns).set_index(X_train.index)
            X_val_scaled = pd.DataFrame(scaler.transform(X_val), 
                                        columns = X_val.columns).set_index(X_val.index)
            
            #Use SMOTETomek to banlance the data
            if smote == True:
                kos = SMOTETomek(random_state=i)  
                X_train_scaled, y_train = kos.fit_resample(X_train_scaled, y_train)
        
            #Fit the model
            model_fit = model.fit(X_train_scaled, y_train)
            
            #Prediction
            y_pred = model.predict(X_val_scaled)
            
            #Get the accurancy and f1 score
            value_train = model_fit.score(X_train_scaled, y_train)
            value_val = model_fit.score(X_val_scaled,y_val)
            value_f1 = f1_score(y_val, y_pred)
        
            score_train.append(value_train)
            score_val.append(value_val)
            score_f1.append(value_f1)
            list_scaler.append(scaler)

    score = pd.Series([np.mean(score_train), np.mean(score_val), np.mean(score_f1)], 
                       index=["Train","Validation","F1_score"])
                       
    score_std = pd.Series([np.std(score_train),np.std(score_val),np.std(score_f1)], 
                      index=["sd_Train","sd_Validation","sd_F1"])

    return (score, score_std)
```

```python
def model_result(n_splits, n_repeats, model):

    result = avg_score_skf(n_splits, n_repeats, X, y, model)
    result_out = avg_score_skf(n_splits, n_repeats, X_out, y_out, model)
    result_s = avg_score_skf(n_splits, n_repeats, X, y, model, smote=True)
    result_s_out = avg_score_skf(n_splits, n_repeats, X_out, y_out, model, smote=True)
    
    s = pd.concat([result[0],result_out[0],result_s[0],result_s_out[0]], axis=1).T
    s.index = ["N","No_Out","S","S+No_out"]
    s_sd = pd.concat([result[1],result_out[1],result_s[1],result_s_out[1]], axis=1).T
    s_sd.index = ["N","No_Out","S","S+No_out"]
    
    return(s, s_sd)
```

```python
def show_result(score, score_std):
    plt.figure(figsize=(len(score),4))
    plt.ylim(0, 1)
    plt.errorbar(y = score, 
                 yerr = score_std, 
                 x=score.index.tolist(),
                 linestyle='None', 
                 marker='x',
                 ms = 12,
                 color = "darkorange",
                 ecolor="darkred",
                 elinewidth=2,
                 capsize=8)
    
    plt.title("F1_Score", fontsize=16)
    plt.xticks(fontsize=14)
    plt.show()
```
