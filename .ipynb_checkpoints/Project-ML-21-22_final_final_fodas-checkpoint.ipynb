```python
#We extract information from "Date" and used them new categorical features
df_train["Date"] = pd.to_datetime(df_train["Date"], format="%d-%b-%y")
df_train["Month"] = df_train["Date"].dt.strftime("%m").astype(int)
df_train["Day"] = df_train["Date"].dt.strftime("%d").astype(int)
df_train["Week"] = df_train["Date"].dt.strftime("%a").astype(str)
df_train["Season"] = pd.cut(df_train["Month"],right=False, bins=4, labels=False)+1
df_train.drop(["Date"], axis=1, inplace=True)
df_train.head(3)
```

## [4.1.Feature Grouping](#M)

```python
#Create a varible that means the average product duration for each page
df_train["Product_Duration/Page"] = (df_train["Product_Duration"]/df_train["Product_Pages"]).replace([np.inf, np.NaN], 0)

#Create a varible that means the average product duration for each page
df_train["AccountMng_Duration/Page"] = (df_train["AccountMng_Duration"]/df_train["AccountMng_Pages"]).replace([np.inf, np.NaN], 0)

#Create a varible that means the average product duration for each page
df_train["FAQ_Duration/Page"] = (df_train["FAQ_Duration"]/df_train["FAQ_Pages"]).replace([np.inf, np.NaN], 0)

df_train["Page_all"] = df_train["AccountMng_Pages"]+df_train["Product_Pages"]+df_train["FAQ_Pages"]

df_train["Duration_all"] = df_train["AccountMng_Duration"]+df_train["Product_Duration"]+df_train["FAQ_Duration"]

df_train["Rate_ratio"] = (df_train["GoogleAnalytics_BounceRate"]/df_train["GoogleAnalytics_ExitRate"]).replace([np.inf, np.NaN], 0)

metric_features = df_train.columns.drop(non_metric_features).tolist()
```

### [4.4.1.IQR](#M)
    
</a>

```python
#Check outliers with 4IQR 
q25 = df_train.quantile(.25)
q75 = df_train.quantile(.75)
iqr = (q75 - q25)

upper_lim = q75 + 4 * iqr
lower_lim = q25 - 4 * iqr

filters = []

for feat in metric_features:
    llim = lower_lim[feat]
    ulim = upper_lim[feat]
    filters.append(df_train[feat].between(llim, ulim, inclusive=True))

filter1 = np.all(filters, axis=0)
df_train[filter1].shape[0]/df_train.shape[0]
```

### [5.1.2.Chi-square](#M)
    
</a>

```python
#Use Chi-square test to filter non metric features
p_chi2 = []

for feat in non_metric_features:
    dfObserved = pd.crosstab(target, df_train[feat]) 
    chi2, p, dof, expected = stats.chi2_contingency(dfObserved.values)
    p_chi2.append(round(p,4))
    
pd.Series(p_chi2, index=df_train[non_metric_features].columns).sort_values(ascending=False)
```

```python
#Drop the featrues when p>0.05
droplist=["Day","Country","Week"]

#Renew the feature lists
df_train = df_train.drop(columns=droplist)
non_metric_features = df_train.columns.drop(metric_features).tolist()
```

## [5.2.Warpper(RFE)](#M)
    
</a>

```python
#Prepare data for feature selection models
X = df_copy[metric_features+non_metric_features]
y = target

#Before using target encoder, change the data type of categorical feature to string
X[non_metric_features] = X[non_metric_features].astype(str, copy=False)

#Split training and test dataset
X_train, X_val, y_train, y_val = train_test_split(X, y, 
                                                  test_size = 0.2, 
                                                  random_state = 50, 
                                                  stratify = target, 
                                                  shuffle = True)

#Power transformation 
power = PowerTransformer().fit(X_train[metric_features])
X_train[metric_features] = power.transform(X_train[metric_features])
X_val[metric_features] = power.transform(X_val[metric_features])

#MinMaxScaler
scaler = MinMaxScaler().fit(X_train[metric_features])
X_train[metric_features] = scaler.transform(X_train[metric_features])
X_val[metric_features] = scaler.transform(X_val[metric_features])

#Target encoding
t_encode = TargetEncoder().fit(X_train[non_metric_features], y_train)
X_train[non_metric_features] = t_encode.transform(X_train[non_metric_features])
X_val[non_metric_features] = t_encode.transform(X_val[non_metric_features])
```

### [5.2.1.Linear Model](#M)
    
</a> 

```python
#Choose how many features we should choose
score_high = 0
num = 0           
for n in range(X_train.shape[1]):
    lr = LogisticRegression()
    rfe = RFE(lr, n+1)
    X_train_rfe = rfe.fit_transform(X_train, y_train)
    X_val_rfe = rfe.transform(X_val)
    lr.fit(X_train_rfe,y_train)
    
    score = f1_score(lr.predict(X_val_rfe),y_val)
    
    if(score>score_high):
        score_high = score
        num = n+1
        
print("Optimum number of features: %d" %num)
print("The score is: %f" % score_high)
```

```python
#Find the feature's ranking by RFE
rfe = RFE(estimator=lr, n_features_to_select=1)
X_rfe = rfe.fit_transform(X_train, y_train)

selected_features = pd.Series(rfe.ranking_, index = X_train.columns)
selected_features.sort_values()
```

```python
#Choose first 10 features
feature_rfe_linear = selected_features.sort_values()[:8].index.tolist()
feature_rfe_linear
```

```python
#Prepare data for feature selection models
X = df_copy[metric_features+non_metric_features]
y = target

#Before using target encoder, change the data type of categorical feature to string
X[non_metric_features] = X[non_metric_features].astype(str, copy=False)

#Split training and test dataset
X_train, X_val, y_train, y_val = train_test_split(X, y, 
                                                  test_size = 0.2, 
                                                  random_state = 10, 
                                                  stratify = target, 
                                                  shuffle = True)

#Power transformation 
power = PowerTransformer().fit(X_train[metric_features])
X_train[metric_features] = power.transform(X_train[metric_features])
X_val[metric_features] = power.transform(X_val[metric_features])

#MinMaxScaler
scaler = MinMaxScaler().fit(X_train[metric_features])
X_train[metric_features] = scaler.transform(X_train[metric_features])
X_val[metric_features] = scaler.transform(X_val[metric_features])

#Target encoding
t_encode = TargetEncoder().fit(X_train[non_metric_features], y_train)
X_train[non_metric_features] = t_encode.transform(X_train[non_metric_features])
X_val[non_metric_features] = t_encode.transform(X_val[non_metric_features])
```

## [6.1.Define functions for running models](#M)
    
</a>

```python
def avg_score_skf(n_splits, n_repeats, X, y, model, smote=False):
    
    score_train = []
    score_val = []
    score_f1 = []
    list_scaler = []
    f1_best = 0.0
    
    #Use Repeat Stratified KFold
    for i in range(n_repeats):
        
        skf = StratifiedKFold(n_splits, shuffle=True, random_state=np.random.randint(1000))
        
        for train_index, val_index in skf.split(X,y):
            
            #Split the training and valiation dataset
            X_train, X_val = X.iloc[train_index], X.iloc[val_index]
            y_train, y_val = y.iloc[train_index], y.iloc[val_index]
            
            #Power Transformation
            power = PowerTransformer().fit(X_train[metric_features_linear])
            X_train[metric_features_linear] = power.transform(X_train[metric_features_linear])
            X_val[metric_features_linear] = power.transform(X_val[metric_features_linear])
            
            #MinMaxScaler
            scaler = MinMaxScaler().fit(X_train[metric_features_linear])
            X_train[metric_features_linear] = scaler.transform(X_train[metric_features_linear])
            X_val[metric_features_linear] = scaler.transform(X_val[metric_features_linear])
            
            #Target encoding
            #X_train[non_metric_features_linear] = X_train[non_metric_features_linear].astype(str, copy=False)
            t_encode = TargetEncoder().fit(X_train[non_metric_features_linear], y_train)
            X_train[non_metric_features_linear] = t_encode.transform(X_train[non_metric_features_linear])
            X_val[non_metric_features_linear] = t_encode.transform(X_val[non_metric_features_linear])
            
            #Use SMOTETomek to banlance the data
            if smote == True:
                kos = SMOTETomek(random_state=np.random.randint(1000))  
                X_train, y_train = kos.fit_resample(X_train, y_train)
        
            #Fit the model
            model_fit = model.fit(X_train, y_train)
            
            #Prediction
            y_pred = model.predict(X_val)
            
            #Get the accurancy and f1 score
            value_train = model_fit.score(X_train, y_train)
            value_val = model_fit.score(X_val,y_val)
            value_f1 = f1_score(y_val, y_pred)
            
            #Store all scores
            score_train.append(value_train)
            score_val.append(value_val)
            score_f1.append(value_f1)
            list_scaler.append(scaler)
    
    #Get all scores' average
    score = pd.Series([np.mean(score_train), np.mean(score_val), np.mean(score_f1)], 
                       index=["Train","Validation","F1_score"])
    
    #Get all scores' standard deviation
    score_std = pd.Series([np.std(score_train),np.std(score_val),np.std(score_f1)], 
                      index=["sd_Train","sd_Validation","sd_F1"])

    return (score, score_std)
```

```python
def avg_score_skf_tree(n_splits, n_repeats, X, y, model, smote=False):
    
    score_train = []
    score_val = []
    score_f1 = []
    list_scaler = []
    f1_best = 0.0
    
    #Use Repeat Stratified KFold
    for i in range(n_repeats):
        
        skf = StratifiedKFold(n_splits, shuffle=True, random_state=np.random.randint(1000))
        
        for train_index, val_index in skf.split(X,y):
            
            #Split the training and valiation dataset
            X_train, X_val = X.iloc[train_index], X.iloc[val_index]
            y_train, y_val = y.iloc[train_index], y.iloc[val_index]
        
            #Fit the model
            model_fit = model.fit(X_train, y_train)
            
            #Prediction
            y_pred = model.predict(X_val)
            
            #Get the accurancy and f1 score
            value_train = model_fit.score(X_train, y_train)
            value_val = model_fit.score(X_val,y_val)
            value_f1 = f1_score(y_val, y_pred)
            
            #Store all scores
            score_train.append(value_train)
            score_val.append(value_val)
            score_f1.append(value_f1)
            list_scaler.append(scaler)
    
    #Get all scores' average
    score = pd.Series([np.mean(score_train), np.mean(score_val), np.mean(score_f1)], 
                       index=["Train","Validation","F1_score"])
    
    #Get all scores' standard deviation
    score_std = pd.Series([np.std(score_train),np.std(score_val),np.std(score_f1)], 
                      index=["sd_Train","sd_Validation","sd_F1"])

    return (score, score_std)
```

```python
def model_result(n_splits, n_repeats, model):
    
    #Run model in different dataset:
        #Normal data
    result = avg_score_skf(n_splits, n_repeats, X_linear, y, model)
        #Data without outliers
    result_out = avg_score_skf(n_splits, n_repeats, X_linear_out, y_out, model)
        #Data with SMOTETomek
    result_s = avg_score_skf(n_splits, n_repeats, X_linear, y, model, smote=True)
        #Data without outliers with SMOTETomek
    result_s_out = avg_score_skf(n_splits, n_repeats, X_linear_out, y_out, model, smote=True)
    
    #Store all results into DataFrame
    s = pd.concat([result[0],result_out[0],result_s[0],result_s_out[0]], axis=1).T
    s.index = ["N","No_Out","S","S+No_out"]
    s_sd = pd.concat([result[1],result_out[1],result_s[1],result_s_out[1]], axis=1).T
    s_sd.index = ["N","No_Out","S","S+No_out"]
    
    return(s, s_sd)
```

```python
def show_result(score, score_std):
    
    #Plot the score with mean and std
    plt.figure(figsize=(len(score),4))
    plt.ylim(0, 1)
    plt.errorbar(y = score, 
                 yerr = score_std, 
                 x=score.index.tolist(),
                 linestyle='None', 
                 marker='x',
                 ms = 12,
                 color = "darkorange",
                 ecolor="darkred",
                 elinewidth=2,
                 capsize=8)
    
    plt.title("F1_Score", fontsize=16)
    plt.xticks(fontsize=14)
    plt.show()
```
